```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df = pd.read_csv('Resources/client_dataset.csv')
df ['subtotal'] = df['unit_price'] * df['qty']
df[['unit_price', 'qty', 'subtotal']].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda weight: weight * 7 if weight > 50 else weight * 10)
df[['unit_price','unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['line_subtotal'] = df['unit_price'] * df['qty']
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda weight: weight * 7 if weight > 50 else weight * 10)
sales_tax_rate = 0.0925
df['total_price'] = (df['line_subtotal'] + df['shipping_price']) * (1 + sales_tax_rate)
df[['line_subtotal', 'shipping_price', 'total_price']].head(3)
```

```python
# Format the data and rename the columns to names suitable for presentation.
data = {
    'Client_ID': [33615, 66037, 46820, 38378, 24741],
    'Units': [164313, 43018, 75768, 73667, 239862],
    'Shipping': [1828985.16, 1395152.33, 1601449.00, 3429455.02, 5126447.83],
    'line_price': [8377308.52, 10259514.79, 9743794.36, 12906550.87, 82268892.02],
    'line_cost': [6175313.91, 7004482.98, 7007191.64, 9634720.98, 45688899.71],
    'line profit': [2201994.61, 3255031.81, 2736602.72, 3271829.89, 36579992.31]
}
df = pd.DataFrame(data)

# Define the money columns. 
money_columns = ['Shipping', 'line_price', 'line_cost', 'line profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    if value >= 1_000_000:
        return f"${value / 1_000_000:.2f}M"
    else:
        return f"${value:,.2f}"
# Apply the currency_format_millions function to only the money columns. 
for col in money_columns:
    df[col] = df[col].apply(currency_format_millions)
# Rename the columns to reflect the change in the money format. 
df.rename(columns={
    'Shipping': 'Shipping (millions)',
    'line_price': 'Total Revenue (millions)',
    'line_cost': 'Total Cost (millions)',
    'line profit': 'Total Profit (millions)'
})
print(df)
```
